```python
#Test of normality
#Using Shapiro-Wilk test to evaluate my data population in order to figure out whether or not my sample data proceeds
#a Gaussian distribution
stat,p = stats.shapiro(df["(%)"]) 
alpha = 0.05
if  p < alpha:
    print("p-value {} and test statistic {}. The data dont'n follow a Gaussian distribution".format(p, stat))
if p > alpha:
    print("p-value {} and test statistic {}. I can't conclude the data do not follow a Gaussian shape".format(p,stat))
```

```python
#Figuring Out Descriptive Statistic Values by School
def describe_and_corr(df,field,variable):
    "df: dataset"
    "field: what categorical variable"
    ""
    stat = round(df.groupby(field).agg({variable: ['count','max','min','mean','std','var']}),4)
    scores = df[variable].to_numpy()
    stat = stat.append(pd.DataFrame([[np.count_nonzero(scores),np.max(scores),np.min(scores),np.mean(scores), np.std(scores), np.var(scores)]], index = ['Total'], columns = stat.columns))
    
    #Computting the correlation between performance in the entrance's exam and the school where studants study
    stat['weighted var'] = stat[(variable,'count')] * stat[(variable,'var')]
    #stat.iloc[-1,-1] == np.nan
    weighted_var = np.sum(stat.iloc[0:-1,-1])/(stat.iloc[-1,0] - 4)
    R2 = 1 - (weighted_var/stat.iloc[-1,-2])
    return (stat,"R2:{}".format(R2))
R2_school = describe_and_corr(df,'School',"(%)")[1]
R2_school
```

```python
def plot_box(dataset,x,xlabel,ylabel,title):
    #plotting a box-splot chart
    sea.set_style('darkgrid', {'axes.grid': True})
    plt.figure()
    box = sea.boxplot(data = dataset, x = x, y ='(%)', palette = "Set3", orient = "v", linewidth = 0.5, saturation= 0.76, fliersize= 0)
    swarm = sea.swarmplot(data = dataset, x = x, y = '(%)', palette = 'Set3', orient = 'v', size = 7,alpha= 0.64, edgecolor= 'grey', linewidth = 0.4)
    plt.xlabel(xlabel,fontsize = 11)
    plt.ylabel(ylabel, fontsize = 11)
    #box.set(xlabel = "Performance (%)", ylabel = "School")
    plt.title(title, fontweight = 'bold')


    #Adjusts
    plt.gcf().set_size_inches(23,8)
    plt.gcf().set_dpi(100)
    plt.xticks(rotation = 35)
    #plt.subplots_adjust(left = 0.4)
plot_box(df, 'School',"School","Performance (%)",'School VS Scores - Boxplot')
```

```python
# 5 largest school
largest_school = list(school_describe.iloc[:-1][school_describe.columns[0]].nlargest(6).index)
largest_school
```

```python
#Reindexing the dataframe
def sample(n,field,group,df):
    'n: sample size'
    'field: variable'
    'groups: level of the nominal variable'
    data = df[['Name','(%)',field]]
    sample = df.copy().sort_values(by = field, ascending = True).set_index([field,"Name"])
    #display(sample)

    #Setting up samples to apply One-way ANOVA
    dataset = {}
    for i in group:
        dataset.setdefault(i,sample.loc[i].sample(n = n, replace = True)['(%)'].values)
    #display(dataset)
    samples = pd.DataFrame(dataset)
    return samples
school_sample = sample(16,'School',largest_school,df)
```

```python
#One-Way ANOVA
F,p = stats.f_oneway(school_sample.iloc[:,0],school_sample.iloc[:,1],school_sample.iloc[:,2],school_sample.iloc[:,3],school_sample.iloc[:,4],school_sample.iloc[:,5])
alpha = 0.05
if p < alpha:
    print("F = {}, p-value = {}. I reject null hypothesis Ho. At least one popullation mean is different from others".format(F,p))
if p > alpha:
     print("F = {}, p-value = {}. I don't have enough evidence to reject  null hypothesisHo. All population means are statistically equal".format(F,p))
```

```python
#Kruskal Test
#It's important to use Kruskal Test to analyze different group means when data considerations for One-way ANOVA were not met even if the power test is lower
F,p = stats.kruskal(school_sample.iloc[:,0],school_sample.iloc[:,1],school_sample.iloc[:,2],school_sample.iloc[:,3],school_sample.iloc[:,4],school_sample.iloc[:,5])
alpha = 0.05
if p < alpha:
    print("F = {}, p-value = {}. I reject null hypothesis Ho. At least one popullation mean is different from others".format(F,p))
if p > alpha:
     print("F = {}, p-value = {}. I don't have enough evidence to reject  null hypothesisHo. All population means are statistically equal".format(F,p))
```

### Measure the income per capita

```python
income = df.copy()[['HM_people_live','avg_income',"(%)","Name"]]
income['min'] = income['avg_income'].apply(lambda x: re.compile(r'(\d+),').findall(x)[0] if x == 'Até R$ 1000,00' else(0 if x == "Maior que R$ 5000,00" else re.compile(r'(\d+),').findall(x)[0])).astype(int)
income['max'] = income['avg_income'].apply(lambda x: re.compile(r'(\d+),').findall(x)[0] if x == 'Maior que R$ 5000,00' else(0 if x == "Até R$ 1000,00" else re.compile(r'(\d+),').findall(x)[1])).astype(int)
income.replace(to_replace = "5 ou mais", value = "5", inplace = True)
income = income.astype({'HM_people_live':int})
income['min'] = round(income['min']/income['HM_people_live'],2)
income['max'] = round(income['max']/income['HM_people_live'],2)
income['min'] = income['min'].apply(lambda x: 'Maior que' if x == 0 else x)
income['max'] = income['max'].apply(lambda x: 'Até' if x == 0 else x)
income['range'] = income['min'].astype(str) + ' --- ' + income['max'].astype(str)
income.loc[income['min'] == '0.00','min'] = 'Até'
display(set(income['range']))
income['percapita'] = income['range'].apply(lambda x: "Maior que 1000" if "Maior" in x else ("Maior que 1000" if "1000.0" in x.split(' --- ')[0] else
                                                                       ("Maior que 1000" if  len(x.split(' --- ')[0]) >= 6  else 'Menor que 1000')))
income
```

### Weighted Socioeconomic Level

## $$\Large WSL = \Sigma_{i = 1}^{n} \alpha_{i}P_{i} \ \ \ , i = 1,2,3,..,n$$
## $$ \Large\alpha_1 + \alpha_2 + \alpha_ 3 + ... + \alpha_n = 1 $$
### Pi is a certain measure used to compound the socio-economic metric
### alpha is its weight

### Plotting the expected composition of a class by using socio-economics metrics 

```python
def pie_plot(df,variable):
    plt.figure()
    sea.set_style('darkgrid', {'axes.grid': True})
    def label_function(val):
        return f'{val / 100 * len(df):.0f}\n{val:.0f}%'
    df.groupby(variable).size().plot(kind='pie', autopct=label_function)

    #Label
    plt.title(variable, fontsize = 12, fontweight = 'bold')
    plt.ylabel("Number Students", fontsize = 10, fontweight = 'bold')


    #setting
    plt.gcf().set_size_inches(9,8)
    plt.gcf().set_dpi(100)
pie_plot(df_classified,"School")
```
